```python
import pandas as pd
from sqlalchemy import create_engine
```

### Put median_price_SDcounty.csv file into dataframe

```python
#Median Housing Price San Diego 
median_price_SDcounty_csv = "datasets/median_price_SDcounty.csv"
median_price_SDcounty_df = pd.read_csv(median_price_SDcounty_csv)
median_price_SDcounty_df.head()
```

```python
### get median house price per zip code
```

```python
median_price_SDcounty_df = median_price_SDcounty_df[['zipcode', 'median', 'latitude', 'longitude']]
median_price_SDcounty_df['zipcode'] = median_price_SDcounty_df['zipcode'].astype(str)
median_price_SDcounty_df.dtypes
median_price_SDcounty_df.head()
```

```python
### Put school_ranking.csv into dataframe
```

```python
school_ranking_csv = "datasets/school_ranking.csv"
school_ranking_df = pd.read_csv(school_ranking_csv)
school_ranking_df.head()
```

```python
##Separate the zipcode from the address and create a separate column for it
school_ranking_df['zipcode'] = school_ranking_df.Address.map(lambda x: ' '.join(x.split(' ')[-1:]))
school_ranking_df['zipcode'] = school_ranking_df['zipcode'].astype(str)
school_ranking_df = school_ranking_df[['Name','Ranking', 'Address','zipcode']]
school_ranking_df.head()
```

```python
schoolRanking_medianPrice_df = pd.merge(school_ranking_df,median_price_SDcounty_df, on='zipcode',how='left')
schoolRanking_medianPrice_df
```

```python
schoolRanking_medianPrice_df.to_csv("datasets/schoolRanking_medianPrice.csv")
```

```python
schoolRanking_medianPrice_df.to_json("datasets/schoolRanking_medianPrice.json", orient="records")
```
